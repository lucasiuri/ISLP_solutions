### 1. For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

#### (a) The sample size $n$ is extremely large, and the number of predictors $p$ is small.

A flexible method should perform better, because the large sample size will allow the method to capture better the relationship of the response with the predictors keeping the variance low while and realizing its natural potential for low bias.

#### (b) The number of predictors $p$ is extremely large, and the number of observations $n$ is small.

A less flexible method is more adequate here, because the low $n$ will make it difficult to a very flexible method to capture the patterns in data, especially since $p$ is large. The high dimensionality (large $p$) makes the problem prone to overfitting, so it is important to reduce the variance through a less flexible method, even though it will introduce bias error.

#### (c) The relationship between the predictors and response is highly non-linear.

In this case a flexible method will perform better because it is necessary to capture the non-linear relation between predictors and response.

#### (d) The variance of the error terms, i.e. $σ^2 = Var(ϵ)$, is extremely high.

In this case, a less flexible method is more suited because it mitigates the variance introduced by the error terms.

### 3. We now revisit the bias-variance decomposition.

#### (a) Provide a sketch of typical (squared) bias, variance, training error, test error, and Bayes (or irreducible) error curves, on a single plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five curves. Make sure to label each one.

#### (b) Explain why each of the five curves has the shape displayed in part (a).

**Bias Error**: is the error introduced because of the inflexibility of the model to capture the patterns in data. It is high for low flexibility and decreases as the flexibility increases.

**Variance Error**: is the error introduced when the model becomes too flexible and overfits the data, incorporating noise as relevant. It is very low when the model is inflexible and increases with the flexxibility.

**Bayes or Irreducible Error**: is the inherent error in data, it is the natural variance that exists in every data set. It is not affected by flexbility and remains constant.

**Training Error**: is the error measured when the predictions are made at the same data observations used for training. When the model is inflexible, this error is high but goes to zero as flexibility increases.

**Test Error**: it is the error measured on different data than the training. If depends on the data linearity, but it generally has an U-shape, beginning at some level for inflexible models (it can already be low for linear data) decreases until some point as flexibility increases and then rises up again, meaning that the training overfitted. It can never be lower than the irreducible error.

### 5. What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

The main advantage is that more flexible approaches can adjust better to the training data, which can result in smaller errors for regression or better accuracy for classification. However, this is not a rule, more flexible methods can overfit on the training data and perform badly on test or real data, it happens when the data is highly linear or when it has too much Bayes error. A less flexible approach should be considered for highly linear data or when the irreducible error is very high.

### 6. Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification (as opposed to a nonparametric approach)? What are its disadvantages?

A parametric method trains the model and returns a function $f(X)$ that approaches the real function of the data, after the training it is possible to have acces to the parameters of the function (thus *parametric*). A non-parametric method develops a model that does not have any parameters.
The main advantage of the parametric approaches lie in the explainability of the model. It is very easy to understand the impact of each predictor in the response, making it very well-suited for inference-oriented problems. The main disadvantage is the low flexibility of these methods, which results in a high bias and may not able to achieve low variance.


#### More complete answer from an LLM

A **parametric** statistical learning method makes an explicit assumption about the **functional form** of $f(X)$. For instance, it assumes $f(X)$ is linear: $f(X) = \beta_0 + \beta_1 X_1 + \dots + \beta_p X_p$. The goal is then simply to **estimate a fixed, finite set of parameters** (the $\beta$'s).

A **non-parametric** method, by contrast, explicitly **avoids making assumptions** about the form of $f(X)$. Instead, it seeks to estimate $f(X)$ that gets as close to the data points as possible without being constrained by a fixed set of parameters. The number of parameters (or complexity) often grows with the size of the dataset $n$.

***

#### Advantages and Disadvantages

| Feature | Parametric Approach | Non-Parametric Approach |
| :--- | :--- | :--- |
| **Advantages** | **1. Interpretability (Inference):** The fixed, simple form (e.g., linear) makes it easy to understand how each predictor affects the response (easy to extract coefficient values). **2. Computational Simplicity:** Estimation involves only a fixed number of parameters, making it computationally faster and requiring less data. | **1. Flexibility (Low Bias):** Can accurately fit a much wider range of functional forms for $f(X)$, including highly non-linear or complex relationships. **2. No Misspecification Risk:** Avoids the risk of choosing an incorrect functional form for $f(X)$. |
| **Disadvantages** | **1. High Bias (Model Misspecification):** If the true function $f(X)$ is complex (e.g., highly non-linear) and the assumed model form (e.g., linear) is too simple, the model will be inaccurate. **2. Limited Flexibility:** Restricted in the range of functions it can approximate. | **1. High Variance (Overfitting):** Since the model can adapt closely to the data, it is highly prone to **overfitting** and modeling the noise, especially with small $n$. **2. Data Requirement:** Requires a very **large sample size ($n$)** to accurately estimate $f(X)$. **3. Low Interpretability:** The complex, flexible form makes it difficult to understand the effect of individual predictors (less suited for inference). |

:-- | :-- | :-- | :-- | :--

4 | 0 | 1 | 2 | Green
5 | −1 | 0 | 1 | Green
6 | 1 | 1 | 1 | Red

Suppose we wish to use this data set to make a prediction for $Y$ when $X1 = X2 = X3 = 0$ using K-nearest neighbors.

#### (a) Compute the Euclidean distance between each observation and the test point, $X1 = X2 = X3 = 0$.

For the first observation:
$d_1 = \sqrt{0^2+3^2+0^2} = \sqrt{9} = 3$

The second:
$d_2 = \sqrt{2^2+0^2+0^2} = \sqrt{4} = 2$

The third:
$d_3 = \sqrt{0^2+1^2+3^2} = \sqrt{10} \approx 3.16$

The fourth:
$d_4 = \sqrt{0^2+1^2+2^2} = \sqrt{5} \approx 2.24$

The fifth:
$d_5 = \sqrt{(-1)^2+0^2+1^2}=\sqrt{2} \approx 1.41$

The last (sixth):
$d_6 = \sqrt{1^2+1^2+1^2}=\sqrt{3} \approx 1.73$

#### (b) What is our prediction with $K = 1$? Why?

The prediction is *green*, because it is the class of the closest neighbor to the test point.

#### (c) What is our prediction with $K = 3$? Why?

The prediction is *red*. Because the three nearest neighbors to our test point are 2 (red), 5 (green) and 6 (red). Although the closest neighbor is green, the majority of the 3-nearest neighbors are red.

#### (d) If the Bayes decision boundary in this problem is highly nonlinear, then would we expect the best value for $K$ to be large or small? Why?

The value for $K$ for highly nonlinear scenarios is expected to be small. Because the larger the value of $K$ gets, less flexible KNN becomes. In this case we need more flexibility, thus a lower $K$.

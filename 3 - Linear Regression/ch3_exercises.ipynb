### 2. Carefully explain the differences between the KNN classifier and KNN regression methods.

A KNN classifier considers the K nearest neighbors to the test observation and the attributes to it the predominant class.
A KNN regressor, instead of just attributing a class it attributes the the average value of the neighbors for the each predictor.
The difference is the nature of the task, a classification returns a class as output while regression returns a continuous value.

### 6. Using (3.4), argue that in the case of simple linear regression, the least squares line always passes through the point  $(\bar{x}, \bar{y})$.



A simple linear regression takes the form

$\^y = \^\beta_1x+\^\beta_0$

Now let's see what it happens when $x = \bar{x}$

$\^y = \^\beta_1\bar{x}+\^\beta_0 = \^\beta_1\bar{x}+ \bar{y} - \^\beta_1\bar{x} = \bar{y}$

So yes, the least squares line always passes through the point $(\bar{x},\bar{y})$.

### 7. It is claimed in the text that in the case of simple linear regression of $Y$ onto $X$, the $R^2$ statistic (3.17) is equal to the square of the correlation between $X$ and $Y$ (3.18). Prove that this is the case. For simplicity, you may assume that $\bar{x} = \bar{y} = 0$.









From equation  (3.17) we have

$R^2 = \frac{\sum_{i=1}^n(y_i-\bar{y})^2-\sum_{i=1}^n(y_i-\^y_i)^2}{\sum_{i=1}^n(y_i-\bar{y})^2}  = \frac{\sum_{i=1}^ny_i^2-\sum_{i=1}^n(y_i-\^y_i)^2}{\sum_{i=1}^ny_i^2}$

But since $\^{y}_i = \^\beta_0 + \^\beta_1x_i$ and in this case we have $\^\beta_0 = 0$ and
$\^\beta_1 = \frac{\sum_{j=1}^nx_jy_j}{\sum_{j=1}^n x_j^2}$ because the means are 0, so we have $\^{y}_i = x_i\frac{\sum_{j=1}^nx_jy_j}{\sum_{j=1}^n x_j^2}$

Then we have $R^2 = \frac{\sum_{i=1}^ny_i^2-\sum_{i=1}^n(y_i-\^y_i)^2}{\sum_{i=1}^ny_i^2} = \frac{\sum_{i=1}^ny_i^2-\sum_{i=1}^ny_i^2-2y_i\^{y}_i+\^y_i^2}{\sum_{i=1}^ny_i^2} = \frac{\sum_{i=1}^ny_i^2-\sum_{i=1}^ny_i^2-2y_i\^\beta_1x_i+(\^\beta_1x_i)^2}{\sum_{i=1}^ny_i^2} = \frac{\sum_{i=1}^ny_i^2-\sum_{i=1}^ny_i^2-2y_i(\frac{\sum_{i=1}^nx_iy_i}{\sum_{i=1}^n x_i^2})x_i+((\frac{\sum_{i=1}^nx_iy_i}{\sum_{i=1}^n x_i^2})x_i)^2}{\sum_{i=1}^ny_i^2} = \frac{\sum_{i=1}^ny_i^2-\sum_{i=1}^ny_i^2-2(\frac{\sum_{i=1}^n(x_iy_i)^2}{\sum_{i=1}^n x_i^2})+\frac{\sum_{i=1}^n(x_iy_i)^2}{\sum_{i=1}^n x_i^2}}{\sum_{i=1}^ny_i^2} = \frac{\sum_{i=1}^ny_i^2-\sum_{i=1}^ny_i^2-(\frac{\sum_{i=1}^n(x_iy_i)^2}{\sum_{i=1}^n x_i^2})}{\sum_{i=1}^ny_i^2} = \frac{\frac{\sum_{i=1}^n(x_iy_i)^2}{\sum_{i=1}^n x_i^2}}{\sum_{i=1}^ny_i^2} = \frac{\sum_{i=1}^n(x_iy_i)^2}{\sum_{i=1}^nx_i^2\sum_{i=1}^n y_i^2}$

Then we have
$\begin{equation}
R^2 = \frac{\sum_{i=1}^n(x_iy_i)^2}{\sum_{i=1}^nx_i^2\sum_{i=1}^n y_i^2}
\end{equation}$

On the other hand, if we take the square of the correlation equation (3.18) we have:

$Cor^2(X,Y) = \frac{(\sum_{i=1}^n[(x_i-\bar{x})(y_i-\bar{y})])^2}{[\sqrt{\sum_{i=1}^n(x_i-\bar{x})^2}\sqrt{\sum_{i=1}^n(y_i-\bar{y})^2}]^2} = \frac{\sum_{i=1}^n(x_iy_i)^2}{[\sqrt{\sum_{i=1}^nx_i^2}\sqrt{\sum_{i=1}^ny_i^2}]^2} = \frac{\sum_{i=1}^n(x_iy_i)^2}{\sum_{i=1}^nx_i^2\sum_{i=1}^n y_i^2}$

Then we have:
$\begin{equation}
Cor^2(X,Y) = \frac{\sum_{i=1}^n(x_iy_i)^2}{\sum_{i=1}^nx_i^2\sum_{i=1}^n y_i^2}
\end{equation}$

We see that equations $(1)$ and $(2)$ are the same, as we wanted to prove.                      $\blacksquare$

### 8. This question involves the use of simple linear regression on the ``Auto`` data set.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm
from ISLP import load_data
from ISLP.models import (ModelSpec as MS, summarize)
import matplotlib.pyplot as plt
```

```python
Auto = load_data('Auto')
Auto
```

#### (a) Use the `sm.OLS()` function to perform a simple linear regression with ``mpg`` as the response and ``horsepower`` as the predictor. Use the ``summarize()`` function to print the results. Comment on the output. For example:
#### i. Is there a relationship between the predictor and the response?
#### ii. How strong is the relationship between the predictor and the response?
#### iii. Is the relationship between the predictor and the response positive or negative?
#### iv. What is the predicted mpg associated with a horsepower of 98? What are the associated 95 % confidence and prediction intervals?

```python
y = Auto['mpg']
X = MS(['horsepower']).fit_transform(Auto)
X
```

```python
model = sm.OLS(y,X)
results = model.fit()
summarize(results)
```

```python
new_X = MS(['horsepower']).fit_transform(pd.DataFrame({'horsepower':[98]}))
new_X
```

```python
new_prediction = results.get_prediction(new_X)
print(f'The predicted mean is: {new_prediction.predicted_mean[0]:.3f}.')
print(f'The 95% confidence interval is: {new_prediction.conf_int(alpha = 0.05)}')
print(f'The 95% prediction interval is: {new_prediction.conf_int(obs = True, alpha = 0.05)}')
```

i. The low *p*-value is an indication that there is a relationship between the response and predictor.

ii and iii. The strength of their relation ship is measured by the negative coefficient $\beta_1= -0.16$ which means that for each added unit in `horsepower`, the `mpg` decreases by 0.16 units. This relationship makes sense because car with more power tend to consume more gas.

iv. The responses are printed on the code.

#### (b) Plot the response and the predictor in a new set of axes ``ax``. Use the ``ax.axline()`` method or the ``abline()`` function defined in the lab to display the least squares regression line.

```python
ax = Auto.plot.scatter('horsepower', 'mpg')
ax.axline(xy1 = (0,results.params['intercept']), slope = results.params['horsepower'], color = 'r',linestyle = '--', linewidth = 3)
ax.set_title('Linear Regression of horsepower on mpg');
```

#### (c) Produce some of diagnostic plots of the least squares regression fit as described in the lab. Comment on any problems you see with the fit.

We begin by plotting the residuals.

```python
ax = plt.subplots(figsize = (8,8))[1]
ax.scatter(results.fittedvalues, results.resid)
ax.set_xlabel('Fitted Value')
ax.set_ylabel('Residual')
ax.set_title('Residual Plot')
ax.axhline(0, c= 'k', ls ='--')
```

Now we see the leverage of each observation.

```python
infl = results.get_influence()
ax = plt.subplots(figsize=(8,8))[1]
ax.scatter(np.arange(X.shape[0]), infl.hat_matrix_diag)
ax.set_xlabel('Index')
ax.set_ylabel('Leverage')
ax.set_title('Leverage of Observations');
```

We can observe from the scatter plot of the predictor and response and also from the residuals plot that the data is not linear, both plots shows a non-linear pattern.

From the leverage plot we see that some observations have a high leverage, which can interfere on the least squares regression fit and make the model worse.

```python
y = Auto['mpg']
predictors = Auto.columns.drop(['mpg'])
X = MS(predictors).fit_transform(Auto)
X
```

```python
model_multiple = sm.OLS(y,X)
results_multiple = model_multiple.fit()
summarize(results_multiple)
```

```python
anova_lm(results, results_multiple)
```

i. The `anova_lm()` function tests the improvement of a model relative to another. The null hypothesis is that the larger model (in thi case `results_multiple`) does not improve the previous model (`results`). The low *p*-value is a strong indicative that the model is improved and thus there is a relationship between the response and the predictors.

ii. Looking at the summary, only the `displacement`, `weight`, `year` and `origin` predictors have a low *p-value*. When we performed a simple linear regression, `horsepower` had a low *p*-value, but now it does not. It means that it explained the response well on its own, but now that other variables are included it loses significance, maybe for collinearity since it has a very high correlation with other predictor like `cylinders`, `displacement` and `weight`. We also saw that it did not have a linear relation with the response.

iii. The positive coefficient of `year` suggests that newer vehicles have higher mileage per gallon (i.e. are more economic).

#### (d) Produce some of diagnostic plots of the linear regression fit as described in the lab. Comment on any problems you see with the fit. Do the residual plots suggest any unusually large outliers? Does the leverage plot identify any observations with unusually high leverage?


```python
ax = plt.subplots(figsize=(8,8))[1]
ax.scatter(results_multiple.fittedvalues, results_multiple.resid)
ax.set_title('Multiple Linear Regression Residuals')
ax.set_xlabel('Fitted Value')
ax.set_ylabel('Residual')
ax.axhline(0, c='k', ls='--');
```

```python
infl = results_multiple.get_influence()
ax = plt.subplots(figsize=(8,8))[1]
ax.scatter(np.arange(X.shape[0]), infl.hat_matrix_diag)
ax.set_title('Leverage of Observations in Multiple Linear Regression')
ax.set_xlabel('Index')
ax.set_ylabel('Leverage');
```

```python
print('The high leverage point is:')
Auto.iloc[np.argmax(infl.hat_matrix_diag)]
```

We can see that residuals decreased compared to the simple linear regression. But we can still observe a pattern, which hints on a non-linearity between the response and the predictors. We also have several observations with high residuals, especially for higher fitted values, meaning that these observations are potential outliers.

The leverage plot shows that when we perform the multiple linear regression the number of observations with high leverage decrease significantly. There is one notable exception that is shown in the code and is very perceptible in the plot. This decrease can be explained by the fact that now we have many predictors, and in order to have high leverage an observation must have unusual values for many if its predictors

#### (f) Try a few different transformations of the variables, such as $\log{X}$, $\sqrt{X}$, $X^2$. Comment on your findings.

```python
predictors = Auto.columns.drop(['mpg', 'acceleration', 'horsepower', 'displacement'])
X = MS(predictors).fit_transform(Auto)
X['log_acceleration'] = np.log(Auto['acceleration'])
X['horsepower_sqrt'] = np.sqrt(Auto['horsepower'])
X['displacement_square'] = Auto['displacement'] ** 2
X
```

```python
model_transformations =  sm.OLS(y,X)
results_transformations = model_transformations.fit()
summarize(results_transformations)
```

```python
anova_lm(results, results_transformations)
```

We performed some transformations (squared ``displacement``, square root of ``horsepower`` and natural logarithm of ``acceleration``) on the predictors with high correlation, but that were non-linear (observed from the scatter plot matrix).
